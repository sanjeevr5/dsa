# Intreval problems

## Can a person attend all the meetings?

```python
def can_attend_all(meetings: list[list[int]]) -> bool:
    meetings.sort(key = lambda x: x[0])
    for index in range(1, len(meetings)):
        if meetings[index-1][1] > meetings[index][0]:
            return False
    return True

assert can_attend_all([]) is True
assert can_attend_all([[1, 2]]) is True
assert can_attend_all([[0, 30], [5, 10], [15, 20]]) is False
assert can_attend_all([[7, 10], [2, 4]]) is True
assert can_attend_all([[1, 3], [3, 5], [5, 8]]) is True
assert can_attend_all([[1, 4], [2, 3]]) is False
assert (
    can_attend_all([[2, 4], [1, 2], [7, 8], [5, 6], [6, 8]]) is False
)
assert can_attend_all([[1, 4], [10, 15], [7, 10]]) is True
assert can_attend_all([[1, 5], [2, 6], [6, 8]]) is False
assert can_attend_all([[0, 1], [1, 2]]) is True

print("All test cases passed!")
```

## Non-overlapping intrevals

```python
def non_overlapping_intervals(arr: list[list[int]]) -> list[list[int]]:
    if not arr:
        return []
    arr.sort(key = lambda x : x[0])
    results = [arr[0]]
    for index in range(1, len(arr)):
        last_start, last_end = results[-1]
        curr_start, curr_end = arr[index]
        if curr_start <= last_end:
            results[-1][1] = max(last_end, curr_end)
        else:
            results.append(arr[index])
    return results

assert non_overlapping_intervals([]) == []
assert non_overlapping_intervals([[1, 3]]) == [[1, 3]]
assert non_overlapping_intervals([[1, 3], [2, 6], [8, 10], [15, 18]]) == [
    [1, 6],
    [8, 10],
    [15, 18],
]
assert non_overlapping_intervals([[1, 4], [4, 5]]) == [[1, 5]]
assert non_overlapping_intervals([[4, 7], [1, 4]]) == [[1, 7]]
assert non_overlapping_intervals([[1, 4], [2, 3]]) == [[1, 4]]
assert non_overlapping_intervals([[1, 2], [3, 4], [5, 6]]) == [[1, 2], [3, 4], [5, 6]]
assert non_overlapping_intervals([[1, 10], [2, 3], [4, 8]]) == [[1, 10]]
assert non_overlapping_intervals([[2, 3], [4, 5], [6, 7], [8, 9], [1, 10]]) == [
    [1, 10]
]
assert non_overlapping_intervals([[1, 5], [2, 3]]) == [[1, 5]]

print("All test cases passed!")
```

## Employee free time

```python
def employee_free_time(meetings: list[list[int]]) -> list[list[int]]:
    if not meetings:
        return []
    non_overlapping_int = non_overlapping_intervals(meetings)
    free = []
    for index in range(1, len(non_overlapping_int)):
        last_start, last_end = non_overlapping_int[index - 1]
        curr_start, curr_end = non_overlapping_int[index]
        if last_end < curr_start:
            free.append([last_end, curr_start])
    return free

assert employee_free_time([]) == []
assert employee_free_time([[1, 2]]) == []
assert employee_free_time([[1, 3], [3, 5]]) == []
assert employee_free_time([[1, 3], [5, 7]]) == [[3, 5]]
assert employee_free_time([[1, 3], [2, 4], [6, 8], [9, 10]]) == [
    [4, 6],
    [8, 9],
]
assert employee_free_time([[1, 4], [2, 3], [3, 6], [8, 9]]) == [[6, 8]]
assert employee_free_time([[1, 4], [4, 6], [6, 8]]) == []
assert employee_free_time([[1, 5], [2, 3], [5, 7], [8, 10]]) == [[7, 8]]
assert employee_free_time([[1, 10]]) == []
assert employee_free_time([[1, 2], [2, 3], [3, 4], [10, 12]]) == [[4, 10]]

print("All test cases passed!")
```
