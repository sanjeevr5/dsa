# Subsequence checker

```python
def is_subsequence(s1: str, s2: str):
    t1 = 0
    t2 = 0
    while t1 < len(s1) and t2 < len(s2):
        if s1[t1] == s2[t2]:
            t1 += 1
        t2 += 1
    return t1 == len(s1)

# Basic true cases
assert is_subsequence("", "") is True
assert is_subsequence("", "abc") is True
assert is_subsequence("a", "a") is True
assert is_subsequence("a", "abc") is True
assert is_subsequence("ace", "abcde") is True
assert is_subsequence("abc", "ahbgdc") is True
assert is_subsequence("gksrek", "geeksforgeeks") is True

# Basic false cases
assert is_subsequence("a", "") is False
assert is_subsequence("ax", "a") is False
assert is_subsequence("aec", "abcde") is False
assert is_subsequence("abc", "acb") is False

# Repeated characters
assert is_subsequence("aa", "aab") is True
assert is_subsequence("aa", "aba") is True
assert is_subsequence("aaa", "aa") is False
assert is_subsequence("abc", "aabbcc") is True
assert is_subsequence("aabb", "ab") is False

# Whole string vs larger string
assert is_subsequence("abc", "abc") is True
assert is_subsequence("abc", "zabxc") is True
assert is_subsequence("xyz", "xxyyzz") is True

# Case sensitivity
assert is_subsequence("abc", "AbC") is False
assert is_subsequence("AbC", "AxxbxxC") is True

# Longer examples
assert is_subsequence("leetcode", "lxxeeyytccooddee") is True
assert is_subsequence("longsubseq", "short") is False

print('All tests passed')
```

# Merge sorted arrays

```python
def merge_sorted_arrays(arr1:list[int], arr2:list[int]):
    n1 = len(arr1)
    n2 = len(arr2)
    t1 = 0
    while t1 < n1:
        if arr1[t1] > arr2[0]:
            arr1[t1], arr2[0] = arr2[0], arr1[t1]
            first_element = arr2[0]
            t2 = 1
            while t2 < n2 and first_element > arr2[t2]:
                arr2[t2 - 1] = arr2[t2]
                t2 += 1
            arr2[t2 - 1] = first_element
        t1 += 1

a1 = [1, 4, 7, 8, 10]
b1 = [2, 3, 9]
merge_sorted_arrays(a1, b1)
assert a1 == [1, 2, 3, 4, 7]
assert b1 == [8, 9, 10]

a2 = [1, 5, 9]
b2 = [2, 3, 7, 11]
merge_sorted_arrays(a2, b2)
assert a2 == [1, 2, 3]
assert b2 == [5, 7, 9, 11]

a3 = [1, 2, 3]
b3 = [4, 5, 6]
merge_sorted_arrays(a3, b3)
assert a3 == [1, 2, 3]
assert b3 == [4, 5, 6]

a4 = [4, 5, 6]
b4 = [1, 2, 3]
merge_sorted_arrays(a4, b4)
assert a4 == [1, 2, 3]
assert b4 == [4, 5, 6]

a5 = [1, 3, 5, 7]
b5 = [2, 4, 6, 8]
merge_sorted_arrays(a5, b5)
assert a5 == [1, 2, 3, 4]
assert b5 == [5, 6, 7, 8]

a6 = [1]
b6 = [0]
merge_sorted_arrays(a6, b6)
assert a6 == [0]
assert b6 == [1]

a7 = [0]
b7 = [1]
merge_sorted_arrays(a7, b7)
assert a7 == [0]
assert b7 == [1]

a8 = [1, 2, 2, 3]
b8 = [2, 2, 4]
merge_sorted_arrays(a8, b8)
assert a8 == [1, 2, 2, 2]
assert b8 == [2, 3, 4]

a9 = [1, 10, 20]
b9 = [2, 3, 4, 5, 6]
merge_sorted_arrays(a9, b9)
assert a9 == [1, 2, 3]
assert b9 == [4, 5, 6, 10, 20]

a10 = [2, 2, 2]
b10 = [2, 2]
merge_sorted_arrays(a10, b10)
assert a10 == [2, 2, 2]
assert b10 == [2, 2]

print("All test cases passed")
```

# Unique intersections

```python
def unique_intersections(arr1: list[int], arr2: list[int]) -> list[int]:
    arr1.sort()
    arr2.sort()
    t1 = 0
    t2 = 0
    n1 = len(arr1)
    n2 = len(arr2)
    intersections = []
    while t1 < n1 and t2 < n2:
        while t1 + 1 < n1 and arr1[t1] == arr1[t1 + 1]:
            t1 += 1
        while t2 + 1 < n2 and arr2[t2] == arr2[t2 + 1]:
            t2 += 1
        if arr1[t1] == arr2[t2]:
            intersections.append(arr1[t1])
            t1 += 1
            t2 += 1
        else:
            t1 += 1
    return intersections

assert unique_intersections([1, 2, 2, 3], [2, 2, 4]) == [2]
assert unique_intersections([1, 2, 3, 4], [2, 4]) == [2, 4]
assert unique_intersections([4, 2, 1, 3], [3, 1]) == [1, 3]
assert unique_intersections([1, 1, 1], [1, 1]) == [1]
assert unique_intersections([1, 2, 3], [4, 5, 6]) == []
assert unique_intersections([], [1, 2, 3]) == []
assert unique_intersections([1, 2, 3], []) == []
assert unique_intersections([], []) == []
assert unique_intersections([1, 2, 2, 2, 3, 3], [2, 2, 3, 3, 3]) == [2, 3]
assert unique_intersections([5, 4, 3, 2, 1], [5, 3, 1]) == [1, 3, 5]
assert unique_intersections([10, 20, 30], [10, 20, 30]) == [10, 20, 30]
assert unique_intersections([1, 3, 5, 7], [2, 4, 6, 8]) == []
assert unique_intersections([0, 0, 1, 1], [0, 1, 2]) == [0, 1]
assert unique_intersections([-3, -2, -1, 0], [-2, 0, 2]) == [-2, 0]
assert unique_intersections([100, 200, 300], [50, 150, 250]) == []

print("All test cases passed!")
```
