# Two sum with duplicates

```python
def two_sum(arr: list[int], sm: int) -> bool:
    arr.sort()
    left = 0
    right = len(arr) - 1
    while left < right:
        while left < right and arr[left] == arr[left + 1]:
            left += 1
        while left < right and arr[right] == arr[right - 1]:
            right -= 1
        curr_tot = arr[left] + arr[right]
        if curr_tot == sm:
            return True
        elif curr_tot < sm:
            left += 1
        else:
            right -= 1
    return False

assert two_sum([2, 7, 11, 15], 9) is True
assert two_sum([2, 7, 11, 15], 26) is True
assert two_sum([2, 7, 11, 15], 100) is False
assert two_sum([1, 2, 3, 4, 5], 9) is True
assert two_sum([1, 2, 3, 4, 5], 10) is False
assert two_sum([3, 3], 6) is True
assert two_sum([3, 3], 5) is False
assert two_sum([1, 2, 3, 3, 4, 5], 6) is True
assert two_sum([1, 1, 1, 1], 2) is True
assert two_sum([5], 5) is False
assert two_sum([5, 5], 10) is True
assert two_sum([], 5) is False
assert two_sum([1], 1) is False
assert two_sum([-1, -2, -3, -4], -5) is True
assert two_sum([-1, 0, 1], 0) is True
assert two_sum([-5, -3, 0, 2, 4], -1) is True
assert two_sum([0, 0], 0) is True
assert two_sum([0, 1, 2], 3) is True
assert two_sum([10, 20, 30, 40], 50) is True
assert two_sum([10, 20, 30, 40], 15) is False
assert two_sum([1, 2, 2, 2, 3], 4) is True
assert two_sum([5, 5, 5, 5, 5], 10) is True
assert two_sum([100, 200, 300], 500) is True
assert two_sum([1, 9, 2, 8, 3, 7], 10) is True
assert two_sum([1, 9, 2, 8, 3, 7], 20) is False
assert two_sum([-10, -5, 0, 5, 10], 0) is True
assert two_sum([-10, -5, 0, 5, 10], -15) is True
assert two_sum([7, 1, 5, 3, 6, 4], 10) is True
assert two_sum([2, 3, 4], 6) is True
assert two_sum([2, 3, 4], 8) is False

print("All test cases passed!")
```

# Target limited 2 sum

```python
def target_limited_2_sum(arr: list[int], sm: int) -> int:
    arr.sort()
    left = 0
    right = len(arr) - 1
    # count = 0
    output = float('-inf')
    while left < right:
        curr_sum = arr[left] + arr[right]
        if curr_sum <= sm:
            output = max(output, curr_sum) # count += right - left since that range consists of all the values that is less than the sum
            left += 1
        else:
            right -= 1
    return output

assert target_limited_2_sum([7, 2, 5, 3], 8) == 8
assert target_limited_2_sum([7, 2, 5, 3], 7) == 7
assert target_limited_2_sum([7, 2, 5, 3], 6) == 5
assert target_limited_2_sum([5, 2, 3, 2, 4, 1], 5) == 5
assert target_limited_2_sum([5, 2, 3, 2, 4, 1], 4) == 4
assert target_limited_2_sum([1, 2, 3, 4], 10) == 7
assert target_limited_2_sum([1, 2, 3, 4], 5) == 5
assert target_limited_2_sum([10, 20], 5) == float("-inf")
assert target_limited_2_sum([1, 1, 1], 3) == 2
assert target_limited_2_sum([1, 1, 1], 1) == float("-inf")
assert target_limited_2_sum([5], 10) == float("-inf")
assert target_limited_2_sum([], 10) == float("-inf")
assert target_limited_2_sum([1], 5) == float("-inf")
assert target_limited_2_sum([1, 9, 2, 8], 10) == 10
assert target_limited_2_sum([1, 9, 2, 8], 11) == 11
assert target_limited_2_sum([1, 9, 2, 8], 9) == 9
assert target_limited_2_sum([-1, -2, -3, -4], -5) == -5
assert target_limited_2_sum([-1, 0, 1], 0) == 0
assert target_limited_2_sum([-5, -3, 0, 2, 4], -1) == -1
assert target_limited_2_sum([0, 0], 0) == 0
assert target_limited_2_sum([0, 1, 2], 3) == 3
assert target_limited_2_sum([10, 20, 30, 40], 50) == 50
assert target_limited_2_sum([10, 20, 30, 40], 49) == 40
assert target_limited_2_sum([10, 20, 30, 40], 15) == float("-inf")
assert target_limited_2_sum([1, 2, 2, 2, 3], 4) == 4
assert target_limited_2_sum([5, 5, 5, 5, 5], 10) == 10
assert target_limited_2_sum([5, 5, 5, 5, 5], 9) == float("-inf")
assert target_limited_2_sum([100, 200, 300], 500) == 500
assert target_limited_2_sum([1, 9, 2, 8, 3, 7], 10) == 10
assert target_limited_2_sum([1, 9, 2, 8, 3, 7], 15) == 15
assert target_limited_2_sum([-10, -5, 0, 5, 10], 0) == 0
assert target_limited_2_sum([-10, -5, 0, 5, 10], -15) == -15
assert target_limited_2_sum([7, 1, 5, 3, 6, 4], 10) == 10
assert target_limited_2_sum([2, 3, 4], 6) == 6
assert target_limited_2_sum([2, 3, 4], 5) == 5
assert target_limited_2_sum([2, 3, 4], 4) == float("-inf")

print("All test cases passed!")
```

# Largest container

```python
def largest_container(arr: list[int]) -> int:
    left = 0
    n = len(arr)
    right = n - 1
    max_water = 0
    while left < right:
        width = right - left
        min_height = min(arr[right], arr[left]) # Use this because it overflows
        curr_water = min_height * width
        if curr_water > max_water:
            max_water = curr_water
        if arr[left] < arr[right]: # Use this because arr is not sorted
            left += 1 # We expect maximum
        else:
            right -= 1
    return max_water

assert largest_container([1, 8, 6, 2, 5, 4, 8, 3, 7]) == 49
assert largest_container([1, 1]) == 1
assert largest_container([4, 3, 2, 1, 4]) == 16
assert largest_container([1, 2, 1]) == 2
assert largest_container([1]) == 0
assert largest_container([]) == 0
assert largest_container([5, 5]) == 5
assert largest_container([2, 3]) == 2
assert largest_container([1, 2, 3, 4, 5]) == 6
assert largest_container([5, 4, 3, 2, 1]) == 6
assert largest_container([3, 3, 3, 3]) == 9
assert largest_container([2, 2, 2, 2, 2]) == 8
assert largest_container([2, 1, 5, 6, 2, 3]) == 10
assert largest_container([3, 4, 1, 2, 2, 4, 1, 3, 2]) == 21
assert largest_container([1, 3, 2, 5, 25, 24, 5]) == 24
assert largest_container([1, 2, 4, 3]) == 4
assert largest_container([0, 0]) == 0
assert largest_container([0, 1, 0, 2, 0]) == 2
assert largest_container([0, 2, 0, 2]) == 4
assert largest_container([1, 8, 6, 2, 5, 4, 8, 25, 7]) == 49
assert largest_container([2, 3, 10, 5, 7, 8, 9]) == 36
assert largest_container([6, 9, 3, 4, 5, 8]) == 32

print("All test cases passed!")
```

# Three sum

```python
def three_sum(arr: list[int], tot: int) -> bool:
    arr.sort()
    n = len(arr)
    for index in range(n - 2):
        left = index + 1
        right = n - 1
        while left < right:
            curr_tot = arr[index] + arr[left] + arr[right]
            if curr_tot == tot:
                return True
            elif curr_tot < tot:
                left += 1
            else:
                right -= 1
    return False

assert three_sum([1, 4, 45, 6, 10, 8], 13) == True
assert three_sum([1, 2, 4, 3, 6, 7], 10) == True
assert three_sum([0, -1, 2, -3, 1], 0) == True
assert three_sum([-1, 0, 1, 2, -1, -4], 0) == True
assert three_sum([1, 2, 3, 4, 5], 50) == False
assert three_sum([1, 2, 7, 12], 0) == False
assert three_sum([5, 5, 5], 20) == False
assert three_sum([1, 1, 1, 2, 2, 3], 6) == True
assert three_sum([1, 1, 1, 1], 3) == True
assert three_sum([1, 1, 1, 1], 4) == False
assert three_sum([2, 2, 2, 2, 2], 6) == True
assert three_sum([-2, 0, 1, 1, 2], 0) == True
assert three_sum([-2, -1, 0, 1, 2, 3], 3) == True
assert three_sum([-5, -3, -1, 2, 4, 6], 0) == True
assert three_sum([], 0) == False
assert three_sum([1], 1) == False
assert three_sum([1, 2], 3) == False
assert three_sum([1, 2, 3], 6) == True
assert three_sum([1, 2, 3], 7) == False
assert three_sum([3, 0, -2, -1, 1, 2], 0) == True
assert three_sum([3, 0, -2, -1, 1, 2], 5) == True
assert three_sum([3, 0, -2, -1, 1, 2], 10) == False
assert three_sum([10, -10, 5, -5, 0], 0) == True

print("All test cases passed!")
```

# Approx three sum

```python
def approx_three_sum(arr: list[int], tot: int) -> bool:
    arr.sort()
    n = len(arr)
    best_tot = float('-inf')
    for index in range(n - 2):
        left = index + 1
        right = n - 1
        while left < right:
            curr_tot = arr[index] + arr[left] + arr[right]
            if curr_tot == tot:
                return curr_tot
            elif curr_tot < tot:
                best_tot = max(best_tot, curr_tot)
                left += 1
            else:
                right -= 1
    return best_tot

assert approx_three_sum([1, 2, 3, 4], 6) == 6
assert approx_three_sum([1, 2, 3, 4], 9) == 9
assert approx_three_sum([0, -1, 2, -3, 1], 0) == 0
assert approx_three_sum([-1, 2, 1, -4], 2) == 2
assert approx_three_sum([1, 2, 4, 8, 16], 10) == 7
assert approx_three_sum([1, 2, 4, 8, 16], 20) == 19
assert approx_three_sum([5, 1, 3, 4, 7], 12) == 12
assert approx_three_sum([5, 1, 3, 4, 7], 11) == 11
assert approx_three_sum([2, 3, 5, 10], 15) == 15
assert approx_three_sum([2, 3, 5, 10], 14) == 10
assert approx_three_sum([-2, 0, 1, 1, 2], 0) == 0
assert approx_three_sum([-2, 0, 1, 1, 2], 1) == 1
assert approx_three_sum([-5, -3, -1, 2, 4, 6], 0) == 0
assert approx_three_sum([-5, -3, -1, 2, 4, 6], -1) == -2
assert approx_three_sum([1, 1, 1, 1], 3) == 3
assert approx_three_sum([1, 1, 1, 1], 4) == 3
assert approx_three_sum([2, 2, 2, 2], 7) == 6
assert approx_three_sum([], 10) == float('-inf')
assert approx_three_sum([1], 10) == float('-inf')
assert approx_three_sum([1, 2], 10) == float('-inf')
assert approx_three_sum([1, 2, 3], 100) == 6
assert approx_three_sum([1, 2, 3], 5) == float('-inf')

print("All test cases passed!")
```
