# Vowel exchange (Case Insensitive)

```python
def vowel_exchange(string: str) -> str:
    if string:
        string: list[str] = list(string)
        leng: int = len(string)
        left: int = 0
        right: int = leng - 1
        vowels: tuple[str] = ('a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U')
        while left < right:
            if string[left] in vowels and string[right] in vowels:
                string[left], string[right] = string[right], string[left]
                left += 1
                right -= 1
            elif string[left] not in vowels:
                left += 1
            elif string[right] not in vowels:
                right -= 1
        return ''.join(string)
    return string

assert vowel_exchange("a") == "a"
assert vowel_exchange("ae") == "ea"
assert vowel_exchange("hello") == "holle"

assert vowel_exchange("education") == "odicatuen"
assert vowel_exchange("banana") == "banana"

assert vowel_exchange("") == ""
assert vowel_exchange("rhythm") == "rhythm"
assert vowel_exchange("aeiou") == "uoiea"

assert vowel_exchange("programming is fun") == "prugrimming as fon"

assert vowel_exchange("ApplE") == "EpplA"

print("All tests passed ✅")
```

# Reverse words

- Strings are immutable so inplace is not possible

```python
def reverse_words(string: str) -> str:
    chars = list(string)
    n = len(chars)
    read = 0
    write = 0
    while read < n and chars[read] == ' ':
        read += 1
    while read < n:
        word_start = write
        while read < n and chars[read] != ' ':
            chars[write] = chars[read]
            read += 1
            write += 1
        chars[word_start:write] = chars[word_start:write][::-1]
        while read < n and chars[read] == ' ':
            read += 1
        if read < n:
            chars[write] = ' '
            write += 1
    return ''.join(chars[:write])


assert reverse_words("hello world") == "olleh dlrow"
assert reverse_words("    hello world") == "olleh dlrow"
assert reverse_words("hello world    ") == "olleh dlrow"
assert reverse_words("hello     world") == "olleh dlrow"
assert reverse_words("  hello   world  ") == "olleh dlrow"
assert reverse_words("hello") == "olleh"
assert reverse_words("   hello   ") == "olleh"
assert reverse_words("a b c") == "a b c"
assert reverse_words("  a    b     c  ") == "a b c"
assert reverse_words("python programming") == "nohtyp gnimmargorp"
assert reverse_words("i am") == "i ma"
assert reverse_words("") == ""
assert reverse_words("     ") == ""
assert reverse_words("  the   quick   brown  ") == "eht kciuq nworb"
assert reverse_words("hello123 world456") == "321olleh 654dlrow"
assert reverse_words("a          b") == "a b"
assert reverse_words("a") == "a"
assert reverse_words("ab") == "ba"

print("All test cases passed!")
```

# Reverse word order

```python
def reverse_word_order(string: str) -> str:
    chars = list(string)[::-1]
    n = len(chars)
    read = 0
    write = 0
    while read < n and chars[read] == ' ':
        read += 1
    while read < n:
        word_start = write
        while read < n and chars[read] != ' ':
            chars[write] = chars[read]
            read += 1
            write += 1
        chars[word_start:write] = chars[word_start:write][::-1]
        while read < n and chars[read] == ' ':
            read += 1
        if read < n:
            chars[write] = " "
            write += 1
    return ''.join(chars[:write])

assert reverse_word_order("hello world") == "world hello"
assert reverse_word_order("  hello   world  ") == "world hello"
assert reverse_word_order("    hello    world") == "world hello"
assert reverse_word_order("hello     world") == "world hello"
assert reverse_word_order("hello world    ") == "world hello"
assert reverse_word_order("hello") == "hello"
assert reverse_word_order("   hello   ") == "hello"
assert reverse_word_order("a b c") == "c b a"
assert reverse_word_order("  a    b     c  ") == "c b a"
assert reverse_word_order("python programming") == "programming python"
assert reverse_word_order("i am") == "am i"
assert reverse_word_order("") == ""
assert reverse_word_order("     ") == ""
assert reverse_word_order("  the   quick   brown  ") == "brown quick the"
assert reverse_word_order("hello123 world456") == "world456 hello123"
assert reverse_word_order("a          b") == "b a"
assert reverse_word_order("a") == "a"
assert reverse_word_order("ab") == "ab"
assert reverse_word_order("one two three four") == "four three two one"
assert reverse_word_order("  first  second  third  ") == "third second first"

print("All test cases passed!")
```
